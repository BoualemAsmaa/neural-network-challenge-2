```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# checking the number of values in each column Attrition and Department
print(attrition_df['Attrition'].value_counts())
print(attrition_df['Department'].value_counts())
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
list_of_columns = ['Education','Age','DistanceFromHome', 'JobSatisfaction', 'OverTime' , 'StockOptionLevel' ,
                   'WorkLifeBalance' , 'YearsAtCompany' , 'YearsSinceLastPromotion' , 'NumCompaniesWorked']

# Create X_df using your selected columns
X_df = attrition_df[list_of_columns]
X_df.head()

# Show the data types for X_df
X_df.dtypes
```

```python
# checking the number of each unique value for OverTime column
X_df['OverTime'].value_counts()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_df,y_df, test_size=0.25, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
label_encoder = LabelEncoder()

# Add new code cells as necessary
X_train['OverTime'] = label_encoder.fit_transform(X_train['OverTime'])
X_test['OverTime'] = label_encoder.transform(X_test['OverTime'])
```

```python
# Printing Overtime count values and data type
overtime_count = X_train['OverTime'].value_counts()
print(overtime_count)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
encoder_Department = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
encoder_Department.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_enc_Department = encoder_Department.transform(y_train[['Department']])
y_test_enc_Department = encoder_Department.transform(y_test[['Department']])
print(y_train_enc_Department)
print(y_test_enc_Department)
```

```python
# Create a OneHotEncoder for the Attrition column
encoder_Attrition = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
encoder_Attrition.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_enc_Attrition = encoder_Attrition.transform(y_train[['Attrition']])
y_test_enc_Attrition = encoder_Attrition.transform(y_test[['Attrition']])
print(y_train_enc_Attrition)
print(y_test_enc_Attrition)
```

```python
print(X_train_scaled is None)
print(y_train_enc_Department is None)
print(y_train_enc_Attrition is None)
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
columns_number = len(X_train.columns)

# Create the input layer
input_Layer = layers.Input(shape=(columns_number,), name='input_features')

# Create at least two shared layers
shared_layer_1 = layers.Dense(units=64, activation='relu', name='shared_layer_1')(input_Layer)
shared_layer_2 = layers.Dense(units=128, activation='relu', name='shared_layer_2')(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
Department_hidden = layers.Dense(32, activation='relu', name='Department_hidden')(shared_layer_2)

# Create the output layer
Department_output = layers.Dense(3, activation='softmax', name='Department_output')(Department_hidden)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
# Create the hidden layer
Attrition_hidden = layers.Dense(32, activation='relu', name='Attrition_hidden')(shared_layer_2)

# Create the output layer
Attrition_output = layers.Dense(2, activation='sigmoid', name='Attrition_output')(Attrition_hidden)

```

```python
# Create the model
model = Model(inputs=input_Layer , outputs=[Department_output , Attrition_output])

# Compile the model
model.compile(optimizer='adam',loss={'Department_output': 'categorical_crossentropy',
              'Attrition_output': 'binary_crossentropy'},
              metrics={'Department_output': 'accuracy'
              , 'Attrition_output': 'accuracy'})


# Summarize the model
model.summary()
```

```python
# Train the model
model_fiting = model.fit(X_train_scaled,{'Department_output': y_train_enc_Department, 'Attrition_output': y_train_enc_Attrition},
    epochs=100,
    batch_size=32,
    validation_split=0.2)
```

```python
# Evaluate the model with the testing data
model_loss, Department_loss, Attrition_loss, Department_accuracy, Attrition_accuracy = model.evaluate(X_test_scaled, [y_test_enc_Department, y_test_enc_Attrition])
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {Department_accuracy}")
print(f"Attrition Accuracy: {Attrition_accuracy}")
```
